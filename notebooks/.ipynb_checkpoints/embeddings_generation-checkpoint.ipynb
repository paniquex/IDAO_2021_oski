```python
from tqdm import tqdm


def generate_embeddings(model, dataloader, DEVICE, config):
    t = tqdm(dataloader)
    model.eval()
    samples2embs = {}
    for batch in t:

        wav_ = batch["img"].to(DEVICE)

        output_dict = model(wav_, return_embeddings=True)

        if config["general"]["SED"]:
            if config["general"]["SED_aggregation"] == "mean":
                preds = output_dict["clipwise_output"].cpu().numpy()
            else:
                preds = output_dict["clipwise_output"].cpu().numpy()
        else:
            preds = torch.argmax(output_dict["softmax_x"], dim=1).cpu().numpy()
        # assert len(batch['sample']) == preds.shape[0]
        for sample, pred in zip(batch['sample'], output_dict['embeddings']):
            if len(pred.shape) == 2:
                samples2embs[sample] = pred.cpu().numpy()
            else:
                if sample not in samples2embs:
                    samples2embs[sample] = [pred.cpu().numpy()]
                else:
                    samples2embs[sample].append(pred.cpu().numpy())
    for sample in samples2embs:
        preds = np.vstack(samples2embs[sample])

    for sample in samples2embs:
        preds = np.vstack(samples2embs[sample])

    return samples2embs
    
```

```python
try:
    os.mkdir(config["general"]["out_path"])
except:
    pass

sample2embs = None

with torch.no_grad():
    for model_name in model_names:
        fold = int(model_name[model_name.find("fold") + 4])
        val_dataset = EmbGenerationDataset(df=train, fold=fold, paths_list=paths,
                                           transform=transforms_train)
        val_dataloader = DataLoader(val_dataset,
                                    **config["validation"]["dataloader"])
        model.load_state_dict(torch.load(os.path.join(config["general"]["out_path"], model_name),
                                    map_location=torch.device(DEVICE))['model_state_dict'])
        if sample2embs is None:
            sample2embs = generate_embeddings(model=model, dataloader=val_dataloader,
                          DEVICE=DEVICE, config=config)
        else:
            sample2embs_new = generate_embeddings(model=model, dataloader=val_dataloader,
                          DEVICE=DEVICE, config=config)
            for sample in sample2embs_new:
                if sample in sample2embs:
                    sample2embs[sample].append(sample2embs_new[sample])
                else:
                    sample2embs[sample] = sample2embs_new[sample]
```
